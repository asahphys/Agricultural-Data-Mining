### Interpretasi Hasil

Distribusi hasil panen pada histogram menunjukkan bahwa data tersebar secara cukup merata tanpa adanya outlier yang ekstrem.
Distribusi hasil panen tidak mengikuti pola normal, melainkan menunjukkan kecenderungan memiliki lebih dari satu puncak, khususnya pada rentang nilai sekitar 35–40 dan 50–55. Pola ini kemungkinan berkaitan dengan adanya kategori hasil panen (rendah, sedang, dan tinggi), di mana data cenderung terkumpul pada beberapa kelompok tertentu.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data_pertanian.csv')

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

variables = ['curah_hujan', 'hasil_panen', 'suhu']
titles = ['Distribusi Curah Hujan', 'Distribusi Hasil Panen', 'Distribusi Suhu']
colors = ['#4C9BE8', '#5DBF8A', '#F4845F']

for i, (var, title, color) in enumerate(zip(variables, titles, colors)):
    axes[i].hist(df[var], bins=15, color=color, edgecolor='white')
    axes[i].set_title(title, fontsize=13)
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Frekuensi')

plt.suptitle('Histogram Variabel Pertanian', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('histogram.png', dpi=150, bbox_inches='tight')
plt.show()
```
